## Neural network model


```python
import numpy as np
import pandas as pd
import keras
from keras.layers import Dense
from keras.models import Sequential
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
columns = df.columns
columns
```

```python
x = df[['Preg', 'Plas', 'Pres', 'skin', 'test',
       'mass', 'pedi', 'age']]
x.head()
```

```python
x.shape
```

```python
y = df['class']
```

We want our model to be three layers (except the input layer):

```python
model = Sequential()
model.add(Dense(units=12, input_dim=8, activation='relu'))
model.add(Dense(units=7, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

We put Dense to be Fully Connected Dense. Unit is the number of nodes.
Since the last layer is the output, units = 1.
And since it is a classification type (diabetes or not), we used sigmoid.
The activation function of the hidden layers may not be the same.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=150, batch_size=10)
```
