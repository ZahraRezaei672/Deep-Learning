# DEEP LEARNING
Deep learning is a subset of machine learning, which includes computer vision (CV) and natural language processing (NLP).
Deep learning uses multilayered neural networks, called deep neural networks, to simulate the complex decision-making power of the human brain.
Deep learning and DNNs (deep neural networks) are being applied to solve complex problems in the real world, like weather prediction, facial recognition, and chatbots, plus perform other types of complex data analysis

## Neural networks and artificial neural networks
Pay attention to the image below :



![human_neuron_anatomy (1)](./Images/2.jpg)




Neural networks, or artificial neural networks, attempt to mimic the human brain through a combination of data inputs, weights, and biases, all acting as silicon neurons. These elements work together to accurately detect, classify, and describe objects within the data.
Deep neural networks consist of multiple layers of interconnected nodes, each building on the previous layer to refine and optimize the prediction or categorization. This progression of computations through the network is called **forward** propagation. The input and output layers of a deep neural network are called visible layers. The input layer is where the deep learning model ingests the data for processing, and the output layer is where the final prediction or classification is made.

Another process called backpropagation uses algorithms, such as gradient descent, to calculate errors in predictions, and then adjusts the weights and biases of the function by moving **backwards** through the layers to train the model. Together, forward propagation and backpropagation enable a neural network to make predictions and correct for any errors . Over time, the algorithm becomes gradually more accurate.

Deep learning requires a tremendous amount of computing power. High-performance graphical processing units (GPUs) are ideal because they can handle a large volume of calculations in multiple cores with copious memory available. Distributed cloud computing might also assist. This level of computing power is necessary to train deep algorithms through deep learning. However, managing multiple GPUs on premises can create a large demand on internal resources and be incredibly costly to scale. For software requirements, most deep learning apps are coded with one of these three learning frameworks: JAX, PyTorch or TensorFlow.

## Types of activation functions

### 1. Sigmoid/Logestic :

![10](./Images/10.png)

```python
x = sp.symbols('x')    # Definition of X symbol
```

```python
sigmoid = 1 / (1 + sp.exp(-x)) # Definition of sigmoid function
sigmoid
```

```python
sigmoid_derivative = sp.diff(sigmoid, x) # derivative of sigmoid function
sigmoid_derivative
```

```python
# Convert to Numpy so we can use it
sigmoid_function = sp.lambdify(x, sigmoid, 'numpy')
sigmoid_derivative_function = sp.lambdify(x, sigmoid_derivative, 'numpy') 
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 400) # Definition of X variable
y = sigmoid_function(x) # Definition of Y variable
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x, y, label='Sigmoid', color='green')
plt.scatter(x[200], y[200], marker='x', c='k')

plt.grid(True)
plt.show()
```

```python
yd = sigmoid_derivative_function(x)
plt.figure(figsize=(6, 3))
plt.plot(x, yd, label='Sigmoid Derivative', color='crimson', linestyle="--")
plt.scatter(x[199], yd[199], marker='x', c='k')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x, y, label='Sigmoid', color='green')
plt.plot(x, yd, label='Sigmoid Derivative', color='crimson', linestyle="--")
plt.legend()
plt.grid(True)
plt.show()
```

#### Sigmoid problems
- gradient vanishing : the slope of the graph becomes close to zero from one place to the next.
- The new W and the old W become equal, which means that the algorithm is finished training (it freezes and does not improve).
- Sigmoid is not symmetrical because its turning point is not in the coordinate center.
- e to the power of one number, it can be very large which slows down the speed.

But it normalizes the data.

### 2. Tanh

![22](./Images/22..jpg)

```python
x = sp.symbols('x') # Definition of X symbol
tanh = sp.tanh(x) # Definition of tanh function
tanh
```

```python
tanh_der = sp.diff(tanh, x) # derivative of tanh function
tanh_der
```

```python
tanh_func = sp.lambdify(x, tanh, 'numpy')
tanh_der_func = sp.lambdify(x, tanh_der, 'numpy')
```

```python
x_vals = np.linspace(-10, 10, 400)
y_vals = tanh_func(x_vals)
yd_vals = tanh_der_func(x_vals)
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_vals, label='tanh', color='blue')
plt.plot(x_vals, yd_vals, label='d[tanh]/dx', color='red', linestyle='--')
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

- This activation function is symmetric.
-  It is exactly between -1 and 1 and its symmetry point is 0,0.
-  It normalizes the data. But it still has the problem of vanishing gradient.

### 3. RELU or Rel

To solve the problem of vanishing gradient, we use this activation function. Actually, it is a polynomial function.



```python
# Definition of RELU function
relu = sp.Piecewise(
     (0, x < 0),
     (x, x >= 0)
 )
relu
```

```python
relu_der = sp.diff(relu, x) # derivative of RELU function
relu_der
```

```python
relu_fun = sp.lambdify(x, relu, 'numpy')
relu_der_fun = sp.lambdify(x, relu_der, 'numpy')
```

```python
y_relu = relu_fun(x_vals)
y_relu_der = relu_der_fun(x_vals)
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_relu, color='blue', label='ReLU')
plt.plot(x_vals, y_relu_der, color='red', linestyle='--', label='d[ReLU]')
plt.legend()
plt.grid(True)
plt.show()
```

- The advantage of this activation function is that the problem of vanishing gradient is solved, but the drawback is that it does not consider the negatives.

### 4. Leaky ReLU
It is a polynomial function,Like ReLU.But X has a coefficient in the negative part called α.
The value of alpha should be less than the slope of the positive part (not reach zero).

```python
x = sp.symbols('x')
```

```python
alpha = 0.01
leaky_relu = sp.Piecewise((alpha * x, x < 0),
                         (x, x >= 0))
leaky_relu
```

```python
leaky_relu_derivative = sp.diff(leaky_relu, x)
leaky_relu_derivative
```

```python
leaky_relu_func = sp.lambdify(x, leaky_relu, 'numpy')
leaky_relu_derivative_func = sp.lambdify(x,
                                         leaky_relu_derivative,
                                         'numpy')
```

```python
x_lrelu = np.linspace(-10, 10, 400)
y_lrelu = leaky_relu_func(x_lrelu)
y_dlrelu = leaky_relu_derivative_func(x_lrelu)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_lrelu, y_lrelu, color='blue',
        label=f'Leaky RelU (α={alpha})')
plt.plot(x_lrelu, y_dlrelu, color='red',
        linestyle='--', label='Leaky RelU Derivative')
plt.title('Leaky RelU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### 5. ELU : exponentional linear unit


```python
x = sp.symbols('x')
alpha = 1
elu = sp.Piecewise((alpha * (sp.exp(x) - 1), x < 0),
                   (x, x >= 0))
elu
```

```python
elu_derivative = sp.diff(elu, x)
elu_derivative
```

```python
elu_func = sp.lambdify(x, elu, 'numpy')
elu_derivative_func = sp.lambdify(x, elu_derivative, 'numpy')
```

```python
x_elu = np.linspace(-10, 10, 400)
y_elu = elu_func(x_elu)
y_delu = elu_derivative_func(x_elu)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_elu, y_elu, color='blue',
        label=f'ELU (α={alpha})')
plt.plot(x_elu, y_delu, color='red',
        linestyle='--', label='ELU Derivative')
plt.title('ELU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
